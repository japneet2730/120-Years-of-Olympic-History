# Data Cleaning
isnull() is used to find the null values and isnull().sum() gives the total count of null values.

fillna() is used to replace the null values.

duplicated() finds the duplicates values and duplicated().sum() gives total count of duplicates in the data.

drop_duplicates() deletes the duplicate values and 'inplace=True' makes the changes permanaent in the dataset.

```python
df.isnull()  #'is' is identity operator
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(),inplace = True)
```

```python
df['Height'].fillna(df['Height'].median(),inplace = True)
df['Weight'].fillna(df['Weight'].median(),inplace = True)
df['Medal'].fillna('No', inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]  
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated()
```

```python
df[df.drop(columns=['ID'], inplace=True)]
```

# Visualisation

```python
M_count = df[df["Sex"] == "M"]["Sex"].count()
F_count = df[df["Sex"] == "F"]["Sex"].count()

print("Number of males - ", M_count)
print("Number of females - ", F_count)
```

```python
M_percent = M_count / (M_count + F_count)
F_percent = F_count / (M_count + F_count)

print("Males % - ", M_percent)
print("Females % - ", F_percent)
```

```python
#A countplot is a plot that shows the number of times an item occurs in a data set
sns.countplot(x = 'Sex', data = df)     
```

```python
summer = df[df['Season'] == 'Summer']
winter = df[df['Season'] == 'Winter']


# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(summer['Age'], bins=20, label='Summer')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Athletes in Summer Games')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(winter['Age'], bins=20, label='Winter', color='orange')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Athletes in Winter Games')
plt.show()
```

```python
# Sex ratio of Male and Female participated from 1896'
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
df["Sex"].value_counts().plot.pie(explode=(0,0.1), shadow=True,autopct='%1.2f%%')
plt.title('Sex ratio of Male and Female participated from 1896')
plt.axis('off')  # Remove axis labels


# Distribution of Medals since 1896
filtered = df[df['Medal'].isin(['Gold','Silver','Bronze'])]
plt.subplot(1,2,2)
filtered['Medal'].value_counts().plot.pie(explode=(0.1,0,0), shadow=True,autopct='%1.2f%%', startangle=0)
plt.title('Distribution of Medals since 1896')
plt.axis('off')  # Remove axis labels

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
