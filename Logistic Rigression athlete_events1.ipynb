```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_athlete_events.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le1.fit(df['Sex'])
df['Sex'] = le1.transform(df['Sex'])
```

```python
le2 = LabelEncoder()
le2.fit(df['NOC'])
df['NOC'] = le2.transform(df['NOC'])
```

```python
le3 = LabelEncoder()
le3.fit(df['Season'])
df['Season'] = le3.transform(df['Season'])
```

```python
le4 = LabelEncoder()
le4.fit(df['Sport'])
df['Sport'] = le4.transform(df['Sport'])
```

```python
df
```

```python
x = df[['Sex','Age','Height','Weight','NOC','Season','Sport']]
y = df['Medal']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(pred,y_test)
acc
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
mod = DecisionTreeClassifier()
```

```python
mod.fit(x_train,y_train)
```

```python
pre1 = mod.predict(x_test)
pre1
```

```python
sc1 = accuracy_score(pre1,y_test)
sc1
```
